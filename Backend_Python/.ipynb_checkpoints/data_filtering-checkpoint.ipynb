```python
q1 = df['balance'].quantile(0.25)
q3 = df['balance'].quantile(0.75)
q1, q3
```

```python
iqr = q3 - q1
iqr
```

```python
upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr
upper_limit, lower_limit
```

```python
if(upper_limit > (df['balance'].max())):
    upper_limit = df['balance'].max()
if(lower_limit < (df['balance'].min())):
    lower_limit = df['balance'].min()
```

```python
upper_limit, lower_limit
```

```python
# treating outliers
def limit_imputer(value):
  if value > upper_limit:
    return upper_limit
  if value < lower_limit:
    return lower_limit
  else:
    return value
```

```python
df['balance'] = df['balance'].apply(limit_imputer)
```

```python
for col in [df]:
  col.loc[ col['balance'] <= 0, 'balance'] = 0
  col.loc[ (col['balance'] > 0) & (col['balance'] <= 61500), 'balance'] = 1
  col.loc[ (col['balance'] > 61500) & (col['balance'] <= 135300), 'balance'] = 2
  col.loc[ (col['balance'] > 135300), 'balance'] = 3
```

```python
upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr
upper_limit, lower_limit
```

```python
if(upper_limit > (df['duration'].max())):
    upper_limit = df['duration'].max()
if(lower_limit < (df['duration'].min())):
    lower_limit = df['duration'].min()
```

```python
upper_limit, lower_limit
```

```python
# treating outliers
def limit_imputer(value):
  if value > upper_limit:
    return upper_limit
  if value < lower_limit:
    return lower_limit
  else:
    return value
```

```python
df['duration'] = df['duration'].apply(limit_imputer)
```
